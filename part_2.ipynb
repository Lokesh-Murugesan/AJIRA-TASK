```python
#check unique values in dataset
df.apply(lambda x: len(x.unique()))
```

```python
#check for categorical attributes
df2 = []
for x in df.dtypes.index:
    if df.dtypes[x] == 'object':
        df2.append(x)
df2
```

```python
#reomove the some attribute
df2.remove('city.1')
df2.remove('name.1')
df2.remove('state.1')
df2.remove('contact.mobile.1')
df2.remove('groupName.1')
df2.remove('bills.date.1')
df2.remove('bills.storeName.1')
df2.remove('bills.products.subCategory.1')
df2.remove('bills.products.name.1')
df2.remove('bills.products.id.1')
df2.remove('bills.products.category.1')
df2.remove('id.1')
df2.remove('bills.isEmi.1')
df2.remove('bills.billNo.1')
df2.remove('bills.isEmi')
df2
```

```python
#print the categorical columns
for col in df2:
    print(col)
    print(df[col].value_counts())
    print()
```

```python
# filling a null values using fillna()
df["city"].fillna("No city", inplace = True)
df["state"].fillna("No state", inplace = True)
df["interests"].fillna("0", inplace = True)
df["name"].fillna("No name", inplace = True)
df["id"].fillna("0", inplace = True)
df["bills.products.category"].fillna("Others", inplace = True)
df["contact.mobile"].fillna("0", inplace = True)
df["bills.date"].fillna("0", inplace = True)
df["bills.storeName"].fillna("Others", inplace = True)
df["bills.billNo"].fillna("0", inplace = True)
df["bills.products.subCategory"].fillna("Others", inplace = True)
df["bills.products.name"].fillna("Others", inplace = True)
df["bills.products.id"].fillna("0", inplace = True)
```

```python
#ploting quantity of item types
l = list(df['bills.storeName'].unique())
chart = sns.countplot(df["bills.storeName"])
chart.set_xticklabels(labels=l, rotation=90)
```
